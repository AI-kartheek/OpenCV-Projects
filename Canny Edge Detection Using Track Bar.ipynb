### Canny Edge Detection
* Canny Edge Detection is a popular edge detection algorithm.

```python
%matplotlib inline
import numpy as np
import cv2 
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Images/messi5.jpg',1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

edges = cv2.Canny(img,100,200)

plt.figure(figsize=(16, 14))
plt.subplot(121),plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges)
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cap = cv2.VideoCapture(0)

cv2.namedWindow('Track Bar')

cv2.createTrackbar('Threshold-I', 'Track Bar', 0, 255, nothing)
cv2.createTrackbar('Threshold-II', 'Track Bar', 0, 255, nothing)

while(1):
    ret, frame = cap.read()
    cv2.imshow('Original Frame', frame)
    
    key = cv2.waitKey(1) & 0xFF
    if (key == 27):
        break
        
    thresh_1= cv2.getTrackbarPos('Threshold-I', 'Track Bar')
    thresh_2= cv2.getTrackbarPos('Threshold-II', 'Track Bar')
    edges = cv2.Canny(frame, thresh_1, thresh_2)
    cv2.imshow('Track Bar', edges)
    
cap.release()
cv2.destroyAllWindows()
```
